```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
```

```python
#Uplaod Team and Player stat files
df_player_stat = pd.read_csv('player_stat.csv')
df_team_stat = pd.read_csv('team_stats.csv')
```

```python
print(df_player_stat.columns)
df_player_stat.head()
```

```python
print(df_player_stat['Metrics'].unique())
df_player_stat_Succ_Raid = df_player_stat[df_player_stat['Metrics'] == ' Successful Raids']
```

```python
df_player_stat_Succ_Raid.head()
df_player_stat_Succ_Raid_wo7 = df_player_stat_Succ_Raid[df_player_stat_Succ_Raid['Season_Name']!='SEASON 7' ]
df_player_stat_Succ_Raid_wo7.head()
```

```python
df_player_sraidt = df_player_stat_Succ_Raid.groupby(['Player_Name']).sum().sort_values(by=['Points'], ascending=False).head(10).reset_index()
df_player_sraid = df_player_stat_Succ_Raid_wo7.groupby(['Player_Name']).sum().sort_values(by=['Points'], ascending=False).head(10).reset_index()
```

```python
def dividedet(row):
    return(row['Points']/row['Total_matches_played'])

df_player_sraid['point_perc'] = df_player_sraid.apply(dividedet, axis=1)
print(df_player_sraid.sort_values(by='point_perc', ascending=False))

df_player_sraidt['point_perc'] = df_player_sraidt.apply(dividedet, axis=1)
print(df_player_sraidt.sort_values(by='point_perc', ascending=False))
```

**From the above results its clearly shows that Pradep Narwal will get more Successfull raids percentage**

```python
df_player_stat_Succ_tack = df_player_stat[df_player_stat['Metrics'] == ' Successful Tackles']
df_player_stat_Succ_tack.head()
df_player_stat_Succ_tack_wo7 = df_player_stat_Succ_tack[df_player_stat_Succ_tack['Season_Name']!='SEASON 7' ]
df_player_stat_Succ_tack_wo7.head()

df_player_stackt = df_player_stat_Succ_tack.groupby(['Player_Name']).sum().sort_values(by=['Points'], ascending=False).head(10).reset_index()
df_player_stack = df_player_stat_Succ_tack_wo7.groupby(['Player_Name']).sum().sort_values(by=['Points'], ascending=False).head(10).reset_index()

def dividedet(row):
    return(row['Points']/row['Total_matches_played'])

df_player_stack['point_perc'] = df_player_stack.apply(dividedet, axis=1)
print(df_player_stack.sort_values(by='point_perc', ascending=False))

df_player_stackt['point_perc'] = df_player_stackt.apply(dividedet, axis=1)
print(df_player_stackt.sort_values(by='point_perc', ascending=False))
```

**From the above results its clearly shows that Manjeet Chillar will get more Successfull tackle percentage**
